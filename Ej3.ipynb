```python
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy.interpolate import interp1d

# from scipy import stats as st
import seaborn as sns
sns.set(color_codes=True)

sns.set_palette(sns.color_palette("Set2"))
sns.set(rc={'figure.figsize':(10,8)}, style="white", color_codes=True)


import warnings
warnings.filterwarnings("ignore")
```

```python
def density_probability(x_param):
    if -math.pi / 2 <= x_param <= math.pi / 2:
        return (13/(12*math.pi)) - (x_param**2/math.pi**3)
    else:
        return 0
    

f3x_values = lambda x_param: density_probability(x_param)

x = np.linspace(-2, 2, 1024)
y = [f3x_values(i) for i in x]

print(y)

plt.ylim(-.1, 0.5)
plt.plot(x,y)
# Grafico de f - Punto A
plt.show()
```

```python
# Hallar F, F-1 y graficar - Punto B

def accumulated_probability(x_param):
    if x_param < -math.pi / 2:
        return 0
    if (-math.pi / 2) <= x_param <= (math.pi / 2):
        return 12/24 + (13 * x_param) / (12 * math.pi) - (x_param ** 3 / (3 * (math.pi ** 3)))
    if x_param > math.pi / 2:
        return 1

F3x_values = lambda x_param: accumulated_probability(x_param)

x = np.linspace(-2, 2, 10000)
y = [F3x_values(i) for i in x]

plt.plot(x,y)
```

```python
## Grafico inversa de F

inverse_accumulated_interpolation = interp1d(y, x)
inverse_accumulated_values = lambda x_param: inverse_accumulated_interpolation(x_param)

range = np.linspace(0, 1, 10000)
values = [inverse_accumulated_values(i) for i in range]
plt.plot(range, values)
```

```python
## Grafico histograma

range = np.random.uniform(0, 1, 10000)
values = [inverse_accumulated_values(i) for i in range]

bins = np.linspace(-3, 3, 30)
plt.hist(values, bins = bins)
plt.title("Histograma")
plt.show()
```
