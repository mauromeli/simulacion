```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [10, 10]
```

```python
from infec_simulation import infec_simulation

MAX_X = 50
MAX_Y = 50
PARTICLES_AMOUNT = 100
FRAMES = 100
FPS = 10
MODEL = "B"
TYPE_OF_MOVING = 3

sim = infec_simulation(PARTICLES_AMOUNT, MAX_X, MAX_Y, FRAMES, FPS, MODEL, TYPE_OF_MOVING)

infections_per_cicle_Model = sim.do_animation(save=True, filename='B33.gif')

print(len(infections_per_cicle_Model))
```

```python
from IPython.display import Image
with open('B33.gif','rb') as file:
    display(Image(file.read()))
```

```python
## Calculo las particulas sanas por ciclo
health_per_cicle_Model = []

for i in range(len(infections_per_cicle_Model)):
    
    health_per_cicle_Model.append(100 - infections_per_cicle_Model[i])

print(infections_per_cicle_Model)
```

## Modelo A

### Caso 1: Todos los individuos de la población se mueven sin restricciones dentro del área.

```python
plt.title('Evolucion de la infeccion', fontsize=20)
plt.axis([0, FRAMES, 0, 100])
plt.xlabel('Ciclos', fontsize=16)
plt.ylabel('Numero de infectados / sanos', fontsize=16)
infected = plt.plot(infections_per_cicle_Model)
health = plt.plot(health_per_cicle_Model)

plt.legend((infected[0], health[0]), ('Particlulas infectadas', 'Particulas sanas'), fontsize=12)

plt.show()
```

### Caso 2: Sólo se mueven los individuos sanos. Una vez que un individuo sano se enferma pasan entre 10 y 20 instantes de tiempo con distribución uniforme para el mismo deje de moverse.

```python
plt.title('Evolucion de la infeccion', fontsize=20)
plt.axis([0, FRAMES, 0, 100])
plt.xlabel('Ciclos', fontsize=16)
plt.ylabel('Numero de infectados / sanos', fontsize=16)
infected = plt.plot(infections_per_cicle_Model)
health = plt.plot(health_per_cicle_Model)

plt.legend((infected[0], health[0]), ('Particlulas infectadas', 'Particulas sanas'), fontsize=12)

plt.show()
```
