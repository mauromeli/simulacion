```python
from simulation import Simulation
import matplotlib.pyplot as plt

MAX_X = 10
MAX_Y = 5
PARTICLES_AMOUNT = 100
FRAMES = 1000
FPS = 100

sim = Simulation(amount_of_circles=PARTICLES_AMOUNT, max_x=MAX_X, max_y=MAX_Y, frames=FRAMES, fps=FPS, wall=True, step_size=0.1)
sim.do_animation(save=True, filename='gas_ej9.gif')

fig2, ax_left_density = plt.subplots()
ax_left_density.plot(list(range(0, FRAMES)), sim.left_density)
plt.savefig('left_density.png')


fig2, ax_right_density = plt.subplots()
ax_right_density.plot(list(range(0, FRAMES)), sim.right_density)
plt.savefig('right_density.png')
```
